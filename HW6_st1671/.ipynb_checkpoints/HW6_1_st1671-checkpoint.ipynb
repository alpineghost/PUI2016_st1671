```python
def canconvert(mydata):
    try :
        float(mydata)
        return True
    except ValueError:
        return False
```

### Fit Unitis vs Energy

```python
X = sm.add_constant(np.log10(bblnrgdata.UnitsTotal))
linmodel = sm.OLS(np.log10(bblnrgdata.Total_energy), X, missing='drop').fit()
```

```python
Y=linmodel.predict()
```

```python
pl.scatter(log10(bblnrgdata.UnitsTotal),log10(bblnrgdata.Total_energy),c = 'royalblue', alpha=0.3,s=20)
pl.plot(log10(bblnrgdata.UnitsTotal),Y, 'r' , label="fit line")
plt.xlabel("Number of Units")
plt.ylabel("Energy consumption(kBtu)")
plt.title("Units vs Energy")
pl.legend(loc="best")
```

```python
data=pd.DataFrame()
x = log10(bblnrgdata.UnitsTotal)
y = log10(bblnrgdata.Total_energy)
data['x']=x
data['x2']=x**2
data['y']=y
data.head(2)
```

```python
linmodel2 = smf.ols(formula='y~x+x2',data=data).fit()
x_pred = np.linspace(0,4.5,1000)
a = pd.DataFrame({"x": x_pred, "x2":x_pred**2})
```

```python
print(linmodel2.summary())
```

```python
pl.scatter(x,y,alpha=0.2,s=10)
pl.plot(x_pred,linmodel2.predict(a), "r-",linewidth=2)
pl.xlim(1,4)
plt.ylabel("Energy consumption(kBtu)")
plt.xlabel("Units total")
plt.title("Energy- Units")
```
