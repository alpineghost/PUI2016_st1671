```python
#from __future__ import print_function
import numpy as np
import pylab as pl
%pylab inline
```

### Normal distribution

```python
mysize = (2000 / np.array(range(1,101)))
normdist = []
for i in range(0,len(mysize)):
    normdist.append(np.random.normal(1,0.1,mysize[i]))
```

```python
normmean=[]
for i in range(0,len(mysize)):
    normmean.append(normdist[i].mean())
```

```python
pl.scatter(mysize, normmean,label = "Mean", alpha = 0.5)
pl.title("Normal Distribution")
pl.ylabel("Mean")
pl.xlabel("Sample Size")
pl.legend(loc='upper right')
```

### Poisson distribution

```python
mysize = (2000 / np.array(range(1,101)))
poisdist = []
for i in range(0,len(mysize)):
    poisdist.append(np.random.poisson(1,mysize[i]))
poismean=[]
for i in range(0,len(mysize)):
    poismean.append(poisdist[i].mean())
```

```python
pl.scatter(mysize, poismean,label = "Mean", alpha = 0.5)
pl.title("Poison Distribution")
pl.ylabel("Mean")
pl.xlabel("Sample Size")
pl.legend(loc='upper right')
```

### Binomial distribution

```python
mysize = (2000 / np.array(range(1,101)))
binodist = []
for i in range(0,len(mysize)):
    binodist.append(np.random.binomial(10,0.1,mysize[i]))
binomean=[]
for i in range(0,len(mysize)):
    binomean.append(binodist[i].mean())
```

```python
pl.scatter(mysize, binomean,label = "Mean", alpha = 0.5)
pl.title("Binomial Distribution")
pl.ylabel("Mean")
pl.xlabel("Sample Size")
pl.legend(loc='upper right')
```

### Chi-squared distribution

```python
chidist = []
for i in range(0,len(mysize)):
    chidist.append(np.random.chisquare(1,mysize[i]))
chimean = []
for i in range(0,len(mysize)):
    chimean.append(chidist[i].mean())
```

```python
pl.scatter(mysize, chimean, label = "Mean", alpha = 0.5)
pl.title("Chi-squared Distribution")
pl.ylabel("Mean")
pl.xlabel("Sample Size")
pl.legend(loc='upper right')
```

### Exponential distribution

```python
expodist = []
for i in range(0,len(mysize)):
    expodist.append(np.random.exponential(1,mysize[i]))
expomean = []
for i in range(0,len(mysize)):
    expomean.append(expodist[i].mean())
```

```python
pl.scatter(mysize, expomean,label = "Mean", alpha = 0.5)
pl.title("Exponential Distribution")
pl.ylabel("Mean")
pl.xlabel("Sample Size")
pl.legend(loc='upper right')
```

We can see from the plot that no matter what distribution the population is, as the sample size <br /> getting biger the sample size will getting closer to the population mean.

```python
sample_mean = normmean+poismean+binomean+chimean+expomean
```

```python
pl.hist(sample_mean, alpha = 0.7, color = 'SteelBlue')
pl.title("Sample Mean Distribution")
pl.xlabel("Mean")
pl.ylabel("Frequency")
pl.show()
```

```python
import numpy as np
from scipy.stats import norm

mu, std = norm.fit(sample_mean)
pl.hist(sample_mean, bins=25, normed=True, alpha = 0.7, color='SteelBlue')
print mu
print std
x = np.linspace(0, 2, 100)
p = norm.pdf(x, mu, std)
pl.plot(x, p, 'r')
pl.title("Normed Sample Mean Distribution")
pl.xlabel("Mean")
pl.ylabel("Frequency")
pl.show()
```
