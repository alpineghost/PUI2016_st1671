```python
import numpy as np
import pandas as pd
import pylab as pl
import statsmodels.api as sm
import scipy.stats
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
%pylab inline
```

```python
import os
import json
```

```python
cityid = '5128581'
```

```python
url = (('http://api.openweathermap.org/data/2.5/forecast/city?id=%s&APPID=%s')%(cityid, os.getenv('weatherAPI')))
```

```python
import urllib2
```

```python
response = urllib2.urlopen(url)
data = response.read()
data = json.loads(data)
```

```python
data[u'list']
```

```python
data1 = data[u'list']
print data1[1][u'main']
```

```python
w_info = pd.DataFrame(columns=['grnd_level', 'humidity', 'pressure', 'sea_level','temp','temp_kf',
                               'temp_max','temp_min'])
for i in range(len(data1)):
    grn = data1[i][u'main'][u'grnd_level']
    hum = data1[i][u'main'][u'humidity']
    pre = data1[i][u'main'][u'pressure']
    sea = data1[i][u'main'][u'sea_level']
    temp = data1[i][u'main'][u'temp']
    tempk = data1[i][u'main'][u'temp_kf']
    tempa = data1[i][u'main'][u'temp_max']
    tempi = data1[i][u'main'][u'temp_min']
    df = pd.DataFrame({'grnd_level': [grn], 'humidity': [hum], 'pressure': [pre], 'sea_level':[sea], 'temp':[temp],
                   'temp_kf':[tempk], 'temp_max':[tempa], 'temp_min':[tempi]})
    w_info = w_info.append(df)
df = df.reset_index()
```

```python
w_info.head()
```

```python
x = np.arange(1,41)
x
w_info.temp_min.min()
w_info.temp_max.max()
```

```python
matplotlib.rcParams.update({'font.size': 6})
```

```python
f, (ax1, ax2, ax3, ax4) = plt.subplots(4, sharex=True)
ax1.plot(x, w_info.grnd_level, label = 'Ground Level')
ax1.plot(x, w_info.sea_level, 'b--', label = 'Sea Level')
ax1.plot(x, w_info.pressure, 'k--', label = 'Pressure')
ax1.set_title('Ground and Sea Level')

ax2.plot(x, w_info.temp, 'r--')
ax2.set_title('Tempreture')

ax3.plot(x, w_info.temp_max, 'g--')
ax3.set_ylim([275, 295])

ax4.plot(x, w_info.temp_min, 'y--')
ax4.set_ylim([275, 295])

```
